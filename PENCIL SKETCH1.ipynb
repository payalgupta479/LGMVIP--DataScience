<b><h1>Task 4 - Image to Pencil Sketch with Python</h1></b>

<b><h2>Importing Libraries</h2><b>

```python
import cv2
import matplotlib.pyplot as plt
```

<b><h2>Loading and Plotting the Original image</h2></b>

```python
img=cv2.imread("hey.png")
plt.figure(figsize=(7,7))
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.title("ORIGINAL IMAGE")
plt.show()
```

<b><h2>Converting Original image to Grayscale</h2></b>

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(7,7))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GRAYSCALE IMAGE")
plt.show()
```

<b><h2>Inverting the Image</h2></b>

```python
img_invert=cv2.bitwise_not(img_gray)
plt.figure(figsize=(7,7))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("INVERTED IMAGE")
plt.show()
```

<b><h2>Smoothing the Image</h2></b>

```python
blurred=cv2.GaussianBlur(img_invert,(21,21),0)
plt.figure(figsize=(7,7))
plt.imshow(blurred,cmap="gray")
plt.axis("off")
plt.title("BLURRED IMAGE")
plt.show()
```

<b><h2>Converting the image into pencil sketch</h2></b>

```python
pencil_sketch=cv2.divide(img_gray,255-blurred,scale=255)
plt.figure(figsize=(7,7))
plt.imshow(pencil_sketch,cmap="gray")
plt.title("FINAL IMAGE")
plt.axis("off")

```
